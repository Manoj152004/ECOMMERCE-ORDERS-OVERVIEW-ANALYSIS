```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r"C:\Users\System -02\Downloads\EXCEL CLEANING _02_ecommerce_orders.csv")
df.drop_duplicates
df.head(1)
```

```python

df['pincode'] = df['pincode'].fillna(000000)
df['pincode'] = df['pincode'].astype(int)

df['discount_pct'] = df['discount_pct'].fillna('0')
df['return_status'] = df['return_status'].fillna(df['return_status'].mode()[0])
```

```python
df['city'] =df['city'].str.title()
df['state'] = df['state'].str.title()

for col in ['order_date','delivery_date']:
    df[col] = pd.to_datetime(df[col],format= 'mixed',dayfirst=True)
```

```python
for i in['unit_price','revenue']:
    df[i] = pd.to_numeric(df[i])
```
